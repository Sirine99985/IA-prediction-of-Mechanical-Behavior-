```python
import pandas as pd
import  sklearn.preprocessing
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

import matplotlib.pyplot as plt

# charger les données à partir d'un fichier Excel
data = pd.read_excel("/content/164.xlsx")
data.head(10)
data.describe()
data.columns
data.shape
num_features =data.select_dtypes(include=['int64','float64']).columns
cat_features =data.select_dtypes(include=['object']).columns
print (cat_features)
data.isnull().sum().sort_values(ascending=False)[:50]

for i in num_features :
  print(i)
  data[i].fillna(data[i].median(), inplace=True)
  scale=LabelEncoder()

  for i in cat_features :
    data[i]=scale.fit_transform (data[i])

    data.head(10)

    x=data.drop([1.4 ,2],axis=1)
    y=data[2.4]

    x=x.values()
    y=y.values()


# supprimer les colonnes non numériques si nécessaire
data = data.select_dtypes(include=[float, int])

# créer une instance de l'objet PCA et spécifier que l'on souhaite garder 2 composantes principales
pca = PCA(n_components=2)

# appliquer l'algorithme PCA aux données
data.columns = data.columns.astype(str)
pca.fit(data)

# afficher les noms des variables d'origine correspondant à chaque composante principale
print("Variables correspondantes aux composantes principales:")
for i, component in enumerate(pca.components_):
    print(f"Composante principale {i+1}: {', '.join(data.columns[component.argsort()[::-1]])}")

# récupérer les composantes principales pour chaque observation
principal_components = pca.transform(data)

# afficher les résultats
print("Composantes principales:")
print(principal_components)
plt.plot(principal_components)
```

Cet algorithme utilise la bibliothèque Pandas et la méthode d'analyse en composantes principales (PCA) de la bibliothèque scikit-learn pour effectuer une analyse en composantes principales sur un fichier Excel.

Voici une explication de chaque étape de l'algorithme:

La première ligne importe la bibliothèque Pandas sous le nom "pd".

La deuxième ligne importe la méthode PCA de la bibliothèque scikit-learn.

La troisième ligne charge les données à partir d'un fichier Excel en utilisant la méthode read_excel de Pandas. Le chemin vers le fichier Excel est spécifié entre les guillemets.

La quatrième ligne supprime les colonnes non numériques (c'est-à-dire les colonnes qui ne contiennent pas de valeurs numériques) du jeu de données. Cela est fait en utilisant la méthode select_dtypes de Pandas en spécifiant les types de données à inclure (float et int).

La cinquième ligne crée une instance de l'objet PCA et spécifie que l'on souhaite garder 2 composantes principales en utilisant l'argument n_components de la méthode PCA.

La sixième ligne applique l'algorithme PCA aux données en utilisant la méthode fit de PCA.

La septième ligne affiche les noms des variables d'origine correspondant à chaque composante principale. Cela est fait en utilisant une boucle for pour parcourir les composantes principales stockées dans l'attribut components_ de l'objet PCA. Pour chaque composante principale, les noms des variables correspondantes sont triés par ordre décroissant de leur contribution à la composante principale à l'aide de la méthode argsort, puis affichés à l'écran.

La huitième ligne utilise la méthode transform de PCA pour récupérer les composantes principales pour chaque observation dans les données d'origine.

La neuvième ligne affiche les résultats de l'analyse en composantes principales à l'aide de la fonction print. Les résultats sont stockés dans la variable principal_components.
